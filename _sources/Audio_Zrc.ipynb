```python
import os
import numpy as np
import librosa
import numpy as np
import scipy
import matplotlib.pyplot as plt
import pandas as pd
```

```python
main_folder_path = '/content/drive/MyDrive/PSD/dataset_audio/TESS Toronto emotional speech set data/'

audio_data_list = []
subfolders = ['OAF_angry',"OAF_disgust",'OAF_Fear','OAF_happy','OAF_neutral','OAF_Pleasant_surprise','OAF_Sad','YAF_angry','YAF_disgust','YAF_fear','YAF_happy','YAF_neutral','YAF_pleasant_surprised','YAF_sad']
```

```python
name = []
mean = []
std = []
median = []
min = []
max = []
mode = []
q1 = []
q3 = []
skew = []
kurt = []
# i = 0
mean_zrc = []
zcr_medians = []
zcr_stds = []
zcr_kurts = []
zcr_skews = []

x=[]
y=[]

rmse_means = []
rmse_medians = []
rmse_stds = []
rmse_kurts = []
rmse_skews = []
```

```python
def get_labels(filename):
    # Misalnya, jika nama file berisi informasi label seperti "cat_01.wav" atau "dog_02.wav"
    label = filename.split("_")[0]
    return label


for subfolder in subfolders:
    folder_path = os.path.join(main_folder_path, subfolder)

    for filename in os.listdir(folder_path):
        name.append(filename)
        if filename.endswith('.wav'):
            audio_path = os.path.join(folder_path, filename)


            audio_data, sample_rate = librosa.load(audio_path, sr=None)
            zcr = librosa.feature.zero_crossing_rate(audio_data)

            zcr_mean = zcr.mean()
            zcr_median = np.median(zcr)
            zcr_std = np.std(zcr)
            zcr_kurt = scipy.stats.kurtosis(zcr.ravel())
            zcr_skew = scipy.stats.skew(zcr.ravel())

            zcr_medians.append(zcr_median)
            zcr_stds.append(zcr_std)
            zcr_kurts.append(zcr_kurt)
            zcr_skews.append(zcr_skew)
            mean_zrc.append(zcr_mean)

            # rmse = librosa.feature.rms(audio_data)
            rmse = np.sqrt(np.mean(np.square(audio_data)))


            rmse_mean = rmse.mean()
            rmse_median = np.median(rmse)
            rmse_std = np.std(rmse)
            rmse_kurt = scipy.stats.kurtosis(rmse.ravel())
            rmse_skew = scipy.stats.skew(rmse.ravel())


            rmse_means.append(rmse_mean)
            rmse_medians.append(rmse_median)
            rmse_stds.append(rmse_std)
            rmse_kurts.append(rmse_kurt)
            rmse_skews.append(rmse_skew)

            label = get_labels(filename)
            y.append(label)

            # Menyimpan fitur-fitur audio ke dalam matriks fitur X
            features = [
                zcr_mean, zcr_median, zcr_std, zcr_kurt, zcr_skew,
                rmse_mean, rmse_median, rmse_std, rmse_kurt, rmse_skew,
                name,mean,std,median,min,max,q1,q3,skew,kurt,
                # Tambahkan fitur lainnya di sini sesuai dengan daftar yang Anda miliki
            ]
            x.append(features)





            audio_data_list.append(audio_data)
```

```python
for audio_data in audio_data_list:
    # Menghitung spektrum frekuensi menggunakan FFT
    fft_result = np.fft.fft(audio_data)
    n_samples = len(audio_data)
    freqs = np.fft.fftfreq(n_samples)
    mean.append(np.mean(freqs))
    std.append(np.std(freqs))
    median.append(np.median(freqs))
    min.append(np.amin(freqs))
    max.append(np.amax(freqs))
    # mode.append(stats.mode(freqs))
    q1.append(np.quantile(freqs,0.25))
    q3.append(np.quantile(freqs,0.75))
    skew.append(scipy.stats.skew(freqs))
    kurt.append(scipy.stats.kurtosis(freqs))
```
